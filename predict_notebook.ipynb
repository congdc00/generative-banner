```python
# FOR SUBMISSION 1
import time # bắt buộc dùng thư viện time để tính
from preprocessing import preprocessing
from predict import Model
import pandas as pd
import os
import csv

def write_time_file(file_path, data):
    """
    Write predicted times to a CSV file.

    Parameters:
    - file_path (str): Đường dẫn tới file CSV cần lưu.
    - data (list): Danh sách chứa thời gian dự đoán cho từng tệp.

    Returns:
    None
    """
    # Mở file CSV để ghi
    with open(file_path, 'w', newline='', encoding='utf-8') as file:
        # Tạo đối tượng writer
        csv_writer = csv.writer(file)

        # Ghi header của file CSV (tên cột)
        csv_writer.writerow(['Filename', 'Predicted Time'])

        # Ghi dữ liệu vào file CSV
        csv_writer.writerows(data)

DATA_PATH = "/data/private"
OUTPUT_PATH = "/results/jupyter_submission1"
df = pd.read_csv(f"{DATA_PATH}/info.csv")
prompts = zip(list(df['caption']), list(df['description']), list(df['moreInfo']))
filenames = list(df['bannerImage'])
model = Model("model_1")
os.mkdir("/results/jupyter_submission1/")
os.mkdir("/results/jupyter_submission1/images/")

all_predicted_time = []
for filename, prompt in zip(filenames, prompts):
    t1 = time.time()
    # ***************Start model prediction******************
    # rewrite the code below depending on your model and methods
    input_ = preprocessing(prompt) # convert prompt into expected input
    
    
    image_1 = model.generate(input_) # model generate
    image_2 = model.enhanced(image_1) # postprocess to expected format result
    image_2.save(f"{OUTPUT_PATH}/{filename}") 
    # ***************End model prediction******************
    t2 = time.time()
    predicted_time = t2 - t1
    all_predicted_time.append((filename, predicted_time))
write_time_file("/results/time_submission1.csv", all_predicted_time) # write to results/time_submission1.csv
```
